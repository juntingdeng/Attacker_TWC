```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lr = [1e-5, 5e-5, 1e-4, 5e-4, 1e-3, 5e-3, 1e-2, 5e-2]
lridx = np.arange(1, len(lr)+1)
acc = [0.2307, 0.3398, 0.4415, 0.7937, 0.9456, 0.7382, 0.5922, 0.0748] 
fig, ax = plt.subplots()

ax.plot(lridx, acc)
ax.set_xlabel("Learning Rate")
ax.set_ylabel("Accuarcy / %")
ax.set_xticklabels(lr)
# plt.xticks(lr)
```
