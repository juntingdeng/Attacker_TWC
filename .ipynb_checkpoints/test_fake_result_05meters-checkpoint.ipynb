```python
import numpy as np
import random
# import cupy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchsummary import summary
import torch.utils.data as Data
import torch.optim as optim
from torch.autograd import Variable
import matplotlib.pyplot as plt
import scipy.io
import csv
import os
import seaborn as sns
from classifier_cnn import addNoise2Batch
from GAN_SDR import ClassifierSDR, Discriminator
```

```python
matFiles25=["pluto1_0.25meters_run2.mat",
    "pluto2_0.25meters_run2.mat",
    "pluto3_0.25meters_run2.mat",
    "AD9082_0.25meters_run2.mat",
    "AD9082_CC2595_amp1_0.25meters_run2.mat",
    "TI_TRF3705_0.25meters.mat",
    "TI_TRF3705_CC2595_amp1_0.25meters.mat",
    "TI_TRF3722_CC2595_amp1_0.25meters.mat",
    "TI_TRF3722_0.25meters.mat",
    "Pluto1RX_9082TX_0.25meters_NoAGC10Gain_3pkthreshold_348samples.mat",
    "Pluto1RX_9082_CC2595_TX_0.25meters_3pkthreshold_582samples.mat"]

matFiles50=["pluto1_0.5meters_run2.mat",
    "pluto2_0.5meters_run2.mat",
    "pluto3_0.5meters_run2.mat",
    "AD9082_0.5meters_run2.mat",
    "AD9082_CC2595_amp1_0.5meters_run2.mat",
    "TI_TRF3705_0.5meters.mat",
    "TI_TRF3705_CC2595_amp1_0.5meters.mat",
    "TI_TRF3722_CC2595_amp1_0.5meters.mat",
    "TI_TRF3722_0.5meters.mat",
    "Pluto1RX_9082TX_0.5meters_NoAGC10Gain_3pkthreshold_403samples.mat",
    "Pluto1RX_9082_CC2595_TX_0.5meters_3pkthreshold_319samples.mat"]
```

```python
def load_allSDR_norm(matFiles, distance=0.25):
    datasets_allSDR=None
    targets_allSDR=None
    samples_perSDR=300
    for matfilei in range(len(matFiles)):
        matfile = matFiles[matfilei]
        datasets_load=scipy.io.loadmat("./data/"+str(distance)+"meters/"+matfile)['packet_equalized_allrecord']
        datasets_load = datasets_load[:samples_perSDR]
        targets_load = matfilei*np.ones(datasets_load.shape[0])
        print("SDR: {}, range: {}~{}".format(matfilei, datasets_load.min(), datasets_load.max()))
        datasets_allSDR = np.concatenate([datasets_allSDR, datasets_load], axis=0) if matfilei!=0 else datasets_load
        targets_allSDR = np.concatenate([targets_allSDR, targets_load], axis=0) if matfilei!=0 else targets_load
    datasets_allSDR_magn = np.max((np.abs(datasets_allSDR.min()), np.abs(datasets_allSDR.max())))
    datasets_allSDR = datasets_allSDR/datasets_allSDR_magn
    print("datasets load all sdr shape: ", datasets_allSDR.shape)
    print("targets load all sdr shape: ", targets_allSDR.shape)
    print("datasets load all sdr and norm range: {}~{}".format(datasets_allSDR.min(), datasets_allSDR.max()))
    return datasets_allSDR, targets_allSDR
```

```python
## Test generator with signals collected at 50 cm: intercepted signals at 50 cm and cooresponding synthesized signals
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("device:{}".format(device))
seed=128
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

nSDR=11
samples_perSDR = 300
datasets_allSDR, targets_allSDR = load_allSDR_norm(matFiles50, distance=0.5)
classifier = ClassifierSDR(nSDR=nSDR)
clf_load=torch.load("./checkpoint2/classifier/11SDR_nosnr_normall_epoch200.cnn", map_location=device)
classifier=clf_load["classifier"]

batch_size=32
for SNR in [5, 10, 15, 20, 25, 30, 35]:
    print("----------------SNR:{}dB----------------".format(SNR))
    path = "./checkpoint2/gan/gan_poly_ray_SDR_normAll_nosnrclf_bn_dup/snr"+str(SNR)+"/plot/ckp/"
    ckps = os.listdir(path)
    datasets_total = None
    targets_total = None
    datasets_total_fake = None
    targets_total_fake = None
    for ckp in ckps:
        if len(ckp.split("_"))<2:
            continue
        SDRlabel = int(ckp.split("_")[0][3:])
        SNR = int(ckp.split("_")[1][3:])
        train_acc = float(ckp.split("_")[3][8:])
        val_acc = float(ckp.split("_")[4][6:])
        # print("\nsdr:{}, snr:{}, tran_acc:{}, val_acc:{}".format(SDRlabel, SNR, train_acc, val_acc))
        ckp_load=torch.load(path+ckp, map_location=device)
        generator = ckp_load["generator"]
        discriminator = ckp_load["discriminator"]
        classifier = ckp_load["classifier"]

        data = datasets_allSDR[SDRlabel*samples_perSDR:(SDRlabel+1)*samples_perSDR]
        if data.shape[0]%batch_size!=0:
            data = np.concatenate([data, data[:int(data.shape[0]%batch_size)]], axis=0)
        # print("data shape:", data.shape)
        
        duplicate=8
        data_dup = np.zeros((data.shape[0]*duplicate, data.shape[1], data.shape[2]))
        for d in range(1, duplicate):
            data_dup[d*data.shape[0]: (d+1)*data.shape[0]] = addNoise2Batch(data, snr=SNR, norm=False, device=device)

        # data = addNoise2Batch(data, snr=SNR, norm=False, device=device)
        data_dup = torch.tensor(data_dup, dtype=torch.float, device=device)
        target_dup = SDRlabel*torch.ones(data_dup.shape[0], device=device)
        datasets_total = data_dup if datasets_total==None else torch.cat([datasets_total, data_dup], dim=0)
        targets_total = target_dup if targets_total==None else torch.cat([targets_total, target_dup])

        realdata_loader = Data.DataLoader(data_dup, batch_size=32, shuffle=True, drop_last=True)
        dataset_fake = torch.zeros((data_dup.shape[0], 2, 800), device=device)
        target_fake = SDRlabel*torch.ones(dataset_fake.shape[0], device=device)
        for batch_idx, (data_batch) in enumerate(realdata_loader): 
            dataset_fake[batch_idx*32:(batch_idx+1)*32],mu, logvar = generator.forward(data_batch, device=device)
        datasets_total_fake = dataset_fake if datasets_total_fake==None else torch.cat([datasets_total_fake, dataset_fake], dim=0)
        targets_total_fake = target_fake if targets_total_fake==None else torch.cat([targets_total_fake, target_fake])


    testdatasets_fake=Data.TensorDataset(datasets_total_fake, targets_total_fake)
    testLoader_fake = Data.DataLoader(testdatasets_fake, batch_size=32, shuffle=True, drop_last=True, generator=torch.manual_seed(seed))
    testdatasets=Data.TensorDataset(datasets_total, targets_total)
    testLoader = Data.DataLoader(testdatasets, batch_size=32, shuffle=True, drop_last=True, generator=torch.manual_seed(seed))

    test_acc=0
    test_acc_matrix = torch.zeros((nSDR, nSDR))
    for batch_idx, (data, labels) in enumerate(testLoader):
        data, labels = data.to(device, dtype=torch.float), labels.to(device,  dtype=torch.int64)
        labels_ = classifier.forward(data)
        labels_ = torch.argmax(labels_, dim=1)
        test_acc+=(sum(labels_==labels)/batch_size)
        for i in range(batch_size):
            label=labels[i]
            test_acc_matrix[label][labels_[i]]+=1
    
    test_acc_fake=0
    test_acc_matrix_fake = torch.zeros((nSDR, nSDR))
    for batch_idx, (data, labels) in enumerate(testLoader_fake):
        data, labels = data.to(device, dtype=torch.float), labels.to(device,  dtype=torch.int64)
        labels_ = classifier.forward(data)
        labels_ = torch.argmax(labels_, dim=1)
        test_acc_fake+=(sum(labels_==labels)/batch_size)
        for i in range(batch_size):
            label=labels[i]
            test_acc_matrix_fake[label][labels_[i]]+=1

    test_acc= test_acc/len(testLoader)
    test_acc_spt_matrix = torch.zeros((nSDR)) 
    for i in range(nSDR):
        sumi = sum(test_acc_matrix[i])
        for j in range(nSDR):
            test_acc_matrix[i][j] = test_acc_matrix[i][j]/sumi
        test_acc_spt_matrix[i] = test_acc_matrix[i][i]
        if torch.abs(sum(test_acc_matrix[i])-1)>1e-4:
            print("Wrong!", sum(test_acc_matrix[i]))
    
    
    test_acc_fake= test_acc_fake/len(testLoader_fake)
    test_acc_spt_matrix_fake = torch.zeros((nSDR)) 
    for i in range(nSDR):
        sumi = sum(test_acc_matrix_fake[i])
        for j in range(nSDR):
            test_acc_matrix_fake[i][j] = test_acc_matrix_fake[i][j]/sumi
        test_acc_spt_matrix_fake[i] = test_acc_matrix_fake[i][i]
        if torch.abs(sum(test_acc_matrix_fake[i])-1)>1e-4:
            print("Wrong!", sum(test_acc_matrix_fake[i]))

    print("test acc: ", np.round(test_acc.detach().cpu().numpy(), decimals=4))
    print("test acc fake: ", np.round(test_acc_fake.detach().cpu().numpy(), decimals=4))
    print("test_matix: ", np.round(test_acc_spt_matrix.detach().cpu().numpy(),decimals=4))
    print("test_matix_fake: ", np.round(test_acc_spt_matrix_fake.detach().cpu().numpy(), decimals=4)) 
```

```python
fig, axs = plt.subplots(1,3, gridspec_kw=dict(width_ratios=[10,10,0.5]), figsize=(10,5),layout="compressed" )

ax1=sns.heatmap(test_acc_matrix, annot=True, cbar=False, ax=axs[0], cmap='Blues',annot_kws={"size": 8},)
ax2=sns.heatmap(test_acc_matrix_fake, annot=True, cbar=False, ax=axs[1], cmap='Blues',annot_kws={"size": 8})

cb=fig.colorbar(axs[0].collections[0], cax=axs[2], label ="False Positive Rate (FPR)")
cb.outline.set_linewidth(0)

cb.set_ticks([0,0.2, 0.4, 0.6,0.8, 1])
ax1.set_title("FPR of Real Signals Intercepted at 50 cm",fontsize=10)
ax2.set_title("FPR of Attacker Synthdized Signals \n using Real Signals Intercepted at 50 cm",fontsize=10)
fig.supylabel('Target Transmitter Index', fontsize=11)
fig.supxlabel('Identified Transmitter Index by Classifier',fontsize=11)
# plt.savefig("../../writing/figure/FPRat50cm.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
